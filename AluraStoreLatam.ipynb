# 4. Productos m谩s y menos vendidos

````python
import matplotlib.pyplot as plt
# funcion para realizar graficos
def grafico_ventas(top, bottom,tienda):
  # Gr谩fico de los m谩s vendidos
  # Visualizar
  plt.figure(figsize=(12, 5))

  # Gr谩fico de los m谩s vendidos
  plt.subplot(1, 2, 1)
  top.plot(kind='bar', color='green')
  plt.title("Top 10 Productos M谩s Vendidos -" + tienda)
  plt.xlabel("Producto")
  plt.ylabel("Cantidad Vendida")
  plt.xticks(rotation=90)

  # Gr谩fico de los menos vendidos
  plt.subplot(1, 2, 2)
  bottom.plot(kind='bar', color='red')
  plt.title("Top 10 Productos Menos Vendidos -" + tienda)
  plt.xlabel("Producto")
  plt.ylabel("Cantidad Vendida")
  plt.xticks(rotation=90)

  plt.tight_layout()
  plt.show()
````

````python


ventas_por_producto = tienda['Producto'].value_counts()

# Mostrar los 10 productos m谩s vendidos
top_10 = ventas_por_producto.head(10)

# Mostrar los 10 productos menos vendidos
bottom_10 = ventas_por_producto.tail(10)

# Mostrar en texto
print(" Productos m谩s vendidos:")
print(top_10)

print("\n Productos menos vendidos:")
print(bottom_10)

grafico_ventas(top_10, bottom_10, "tienda 1")
````
