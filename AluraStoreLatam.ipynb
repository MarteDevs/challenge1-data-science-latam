# 4. Productos más y menos vendidos

````python
import matplotlib.pyplot as plt
# funcion para realizar graficos
def grafico_ventas(top, bottom,tienda):
  # Gráfico de los más vendidos
  # Visualizar
  plt.figure(figsize=(12, 5))

  # Gráfico de los más vendidos
  plt.subplot(1, 2, 1)
  top.plot(kind='bar', color='green')
  plt.title("Top 10 Productos Más Vendidos -" + tienda)
  plt.xlabel("Producto")
  plt.ylabel("Cantidad Vendida")
  plt.xticks(rotation=90)

  # Gráfico de los menos vendidos
  plt.subplot(1, 2, 2)
  bottom.plot(kind='bar', color='red')
  plt.title("Top 10 Productos Menos Vendidos -" + tienda)
  plt.xlabel("Producto")
  plt.ylabel("Cantidad Vendida")
  plt.xticks(rotation=90)

  plt.tight_layout()
  plt.show()
````

````python


ventas_por_producto = tienda['Producto'].value_counts()

# Mostrar los 10 productos más vendidos
top_10 = ventas_por_producto.head(10)

# Mostrar los 10 productos menos vendidos
bottom_10 = ventas_por_producto.tail(10)

# Mostrar en texto
print("🔝 Productos más vendidos:")
print(top_10)

print("\n🔽 Productos menos vendidos:")
print(bottom_10)

grafico_ventas(top_10, bottom_10, "tienda 1")
````

````python

# TIENDA 2
ventas_por_producto2 = tienda2['Producto'].value_counts()

# Mostrar los 10 productos más vendidos
top_10_2 = ventas_por_producto2.head(10)

# Mostrar los 10 productos menos vendidos
bottom_10_2 = ventas_por_producto2.tail(10)

# Mostrar en texto
print("🔝 Productos más vendidos:")
print(top_10_2)

print("\n🔽 Productos menos vendidos:")
print(bottom_10_2)

grafico_ventas(top_10_2, bottom_10_2, "Tienda 2")
````

````python
# TIENDA 3
ventas_por_producto3 = tienda3['Producto'].value_counts()

# Mostrar los 10 productos más vendidos
top_10_3 = ventas_por_producto3.head(10)

# Mostrar los 10 productos menos vendidos
bottom_10_3 = ventas_por_producto3.tail(10)

# Mostrar en texto
print("🔝 Productos más vendidos:")
print(top_10_3)

print("\n🔽 Productos menos vendidos:")
print(bottom_10_3)

grafico_ventas(top_10_3, bottom_10_3, "Tienda 3")
````
